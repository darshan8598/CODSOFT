# checking for outliers

```python
numcol=df.select_dtypes(['int64','float64']).columns
```

```python
plt.figure(figsize=(30,30))
plt.suptitle('checking outliers')
pltn=1
for i in numcol:
    if pltn <=30:
        ax=plt.subplot(30,2,pltn)
        sns.boxplot(df[i])
        plt.xlabel(i,fontsize=15)
        pltn=pltn+1
plt.tight_layout()
    
```

```python
#create function to remove outliers\
def impute_outlier(value):
    if value > ul:
        return a
    elif value< ll:
        return a
    else:
        return value
```

```python
for i in df.columns:
    print('------',i,'-------')
    print(df[i].value_counts())
    print('-------------------')
    
```

```python
plt.figure(figsize=(12,15))
plt.suptitle('checking skewness')
pltn=1
for i in numcol:
    if pltn <=30:
        ax=plt.subplot(30,2,pltn)
        sns.distplot(df[i])
        plt.xlabel(i,fontsize=15)
        pltn=pltn+1
plt.tight_layout()
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
#to remove outliers
for i in x.columns:
    a=x[i].mode()[0]
    q1,q3=np.percentile(x[i],[25,75])
    iqr=q3-q1
    ul=q3+1.5*iqr
    ll=q1-15*iqr
    
    x[i]=x[i].apply(impute_outlier)
```

```python
plt.figure(figsize=(30,30))
plt.suptitle('checking outliers')
pltn=1
for i in numcol:
    if pltn <=30:
        ax=plt.subplot(30,2,pltn)
        sns.boxplot(df[i])
        plt.xlabel(i,fontsize=15)
        pltn=pltn+1
plt.tight_layout()
```
